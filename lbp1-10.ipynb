### Test #2 Program

Write a program to read three values and perform an addition operation.

#### C Implementation


## LBP1 | Even or Odd Number

### LBP1: Even or Odd

* [LBP01](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp1)

Write a program that determines whether a given integer is even or odd.

**Requirements:**

* Input: A single integer, `n`
* Constraints: `n ≥ 0`
* Output:
  * Print `"even"` if `n` is even.
  * Print `"odd"` if `n` is odd.
  * Print `"invalid"` if the input does not satisfy the constraint.

#### Java Implementation

```java
import java.util.Scanner;

public class Main {
  public static void main(String... args) {
    Scanner sc = new Scanner(System.in);
    int number = sc.nextInt();

    if (number < 0) {
      System.out.println("invalid");
    } else {
      String result = ((number & 1) == 1) ? "odd" : "even";
      System.out.println(result);
    }
    sc.close();
  }
}
```

#### Python Implementation

````python
"""
The bitwise AND `&` operation checks the least significant bit (LSB):
- For odd numbers, the LSB is 1 (n & 1 == 1)
- For even numbers, the LSB is 0 (n & 1 == 0)
NOTE: `n & 0` will always result in 0, and hence it can't be used to determine even or odd number.
"""
def even_odd(n: int) -> str:
  return "invalid" if n < 0 else "even" if n % 2 == 0 else "odd"

n: int = int(input())
even_odd(n)
````

````python
# using divisor method
# dividing by 2 and multiplying by 2 gives back the original number
even_odd = lambda n:"invalid" if n < 0 else "even" if (n // 2) * 2 == n else "odd"

n: int = int(input())
even_odd(n)
````

````python
# using Bitwise `&` operator
def even_odd(n: int) -> str:
  return "invalid" if n < 0 else "even" if n & 1 == 0 else "odd"

n: int = int(input())
even_odd(n)
````

## LBP2 | Weird or Not Weird

### LBP2: Weird or Not Weird

* [LBP02](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp2)
* [Python If-Else | HackerRank](https://www.hackerrank.com/challenges/py-if-else/problem)

You are given an integer `n` and need to check several conditions to print either `"Weird"` or `"Not Weird"`.

* If the integer `n` is odd, it is classified as `"Weird"`.
* If `n` is even and falls within the range `2 to 5`, it is classified as `"Not Weird"` .
* If `n` is even and falls within the range `6 to 20`, it is classified as `"Weird"`.
* If `n` is even and greater than `20`, it is classified as `"Not Weird"`.

**Requirements:**

* Input: A single integer `n`
* Constraints: `1 ≤ n ≤ 100`
* Output: `"Weird"` or `"Not Weird"` based on the conditions.

#### C Implementation

```c
#include <stdio.h>

char *weirdNotWeird(int n) {
  if (n % 2 == 1) {
    return "Weird";
  } else {
    if ((n >= 2 && n <= 5) || (n > 20)) {
      return "Not Weird";
    } else if (n >= 6 && n <= 20) {
      return "Weird";
    } 
  }
  // to avoid compilation error
  return "";
}

int main() {
  int n;
  scanf("%d", &n);

  printf("%s", weirdNotWeird(n));
  return 0;
}
```

#### Python Implementation

````python
def weird_notWeird(n: int) -> str:
  if n & 1 == 1:
    return "Weird"
  else:
    if 2 <= n <= 5 or n > 20:
      return "Not Weird"
    elif 6 <= n <= 20:
      return "Weird"

n: int = int(input())
weird_notWeird(n)
````
