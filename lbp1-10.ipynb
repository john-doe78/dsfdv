## LBP1 | Even or Odd Number

### LBP1: Even or Odd

* [LBP01](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp1)

Write a program that determines whether a given integer is even or odd.

**Requirements:**

* Input: A single integer, `n`
* Constraints: `n ≥ 0`
* Output:
  * Print `"even"` if `n` is even.
  * Print `"odd"` if `n` is odd.
  * Print `"invalid"` if the input does not satisfy the constraint.

#### Java Implementation

```java
import java.util.Scanner;

public class Main {
  public static void main(String... args) {
    Scanner sc = new Scanner(System.in);
    int number = sc.nextInt();

    if (number < 0) {
      System.out.println("invalid");
    } else {
      String result = ((number & 1) == 1) ? "odd" : "even";
      System.out.println(result);
    }
    sc.close();
  }
}
```

#### Python Implementation

````python
"""
The bitwise AND `&` operation checks the least significant bit (LSB):
- For odd numbers, the LSB is 1 (n & 1 == 1)
- For even numbers, the LSB is 0 (n & 1 == 0)
NOTE: `n & 0` will always result in 0, and hence it can't be used to determine even or odd number.
"""
def even_odd(n: int) -> str:
  return "invalid" if n < 0 else "even" if n % 2 == 0 else "odd"

n: int = 12
even_odd(n)
````

````python
# using divisor method
# dividing by 2 and multiplying by 2 gives back the original number
even_odd = lambda n:"invalid" if n < 0 else "even" if (n // 2) * 2 == n else "odd"

n: int = 13
even_odd(n)
````

````python
# using Bitwise `&` operator
def even_odd(n: int) -> str:
  return "invalid" if n < 0 else "even" if n & 1 == 0 else "odd"

n: int = -12
even_odd(n)
````

## LBP2 | Weird or Not Weird

### LBP2: Weird or Not Weird

* [LBP02](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp2)
* [Python If-Else | HackerRank](https://www.hackerrank.com/challenges/py-if-else/problem)

You are given an integer `n` and need to check several conditions to print either `"Weird"` or `"Not Weird"`.

* If the integer `n` is odd, it is classified as `"Weird"`.
* If `n` is even and falls within the range `2 to 5`, it is classified as `"Not Weird"` .
* If `n` is even and falls within the range `6 to 20`, it is classified as `"Weird"`.
* If `n` is even and greater than `20`, it is classified as `"Not Weird"`.

**Requirements:**

* Input: A single integer `n`
* Constraints: `1 ≤ n ≤ 100`
* Output: `"Weird"` or `"Not Weird"` based on the conditions.

#### C Implementation

```c
#include <stdio.h>

char *weirdNotWeird(int n) {
  if (n % 2 == 1) {
    return "Weird";
  } else {
    if ((n >= 2 && n <= 5) || (n > 20)) {
      return "Not Weird";
    } else if (n >= 6 && n <= 20) {
      return "Weird";
    } 
  }
  // to avoid compilation error
  return "";
}

int main() {
  int n;
  scanf("%d", &n);

  printf("%s\n", weirdNotWeird(n));
  return 0;
}
```

#### Python Implementation

````python
def weird_notWeird(n: int) -> str:
  if n & 1 == 1:
    return "Weird"
  else:
    if (2 <= n <= 5) or (n > 20):
      return "Not Weird"
    elif (6 <= n <= 20):
      return "Weird"
````

## LBP3-4 | Leap Year, E-commerce Bookshelf

### LBP3: Leap Year or Not

* [LBP03](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp3)
* [Leap Year | HacerRank](https://www.hackerrank.com/challenges/write-a-function/problem)

Given an integer representing a year, determine whether the year is a leap year.

A year is a leap year if:

* It is divisible by 400, or
* It is divisible by 4 but not divisible by 100.

**Requirements:**

* Input: An integer `n`.
* Constraints: `1900 ≤ year ≤ 10⁵`.
* Output: `"True"` if the given year is a leap year, `"False"` otherwise.

#### Java Implementation

```java
import java.time.Year;
import java.util.Scanner;

public class Main {
  public static void main(String... args) {
    Scanner sc = new Scanner(System.in);
    int year = sc.nextInt();
    
    String result = (Year.isLeap(year)) ? "True" : "False";
    System.out.println(result);
    sc.close();
  }
}
```

#### C Implementation

```c
#include <stdio.h>

char *isLeap(int year) {
  if ((year % 400 == 0 && year % 100 == 0) || (year % 100 != 0 && year % 4 == 0)) {
    return "True";
  } else {
    return "False";
  }
}

int main() {
  int year;
  scanf("%d", &year);
  
  char *result = isLeap(year);
  printf("%s\n", result);
  return 0;
}
```

#### Python Implementation

````python
"""
A year is a leap year only if it meets the following conditions:
- It is not a century year (1000, 1900, 2000) and is divisible by 4.
- If it is a century year, it must be divisible by 400.
"""
import calendar

def is_leap(year: int) -> str:
  return "True" if calendar.isleap(year) else "False"

year: int = 1600
is_leap(year)
````

### LBP4: E-commerce Bookshelf

* [LBP04](https://www.hackerrank.com/contests/41dec30av2022/challenges/lbp4)

The e-commerce company Bookshelf tracks its monthly book sales, which range from 30 to 100 units. To manage inventory efficiently, the company divides sales into four categories. Each category helps decide whether to increase or decrease stock for the following month.

The sales groups are defined as follows:

| Sales Range | Group |
|-------------|-------|
| 30-50       | D     |
| 51-60       | C     |
| 61-80       | B     |
| 81-100      | A     |

Write an algorithm to determine the appropriate group based on the number of books sold in a month.

**Requirements:**

* Input: A number entered by user.
* Constraints: `30 ≤ sales ≤ 100`
* Output: A character representing the sales group.

#### Python Implementation

````python
def bookshelf_group(sales: int) -> str:
  if 30 <= sales <= 50: return "D"
  elif 51 <= sales <= 60: return "C"
  elif 61 <= sales <= 80: return "B"
  elif 81 <= sales <= 100: return "A"
  else: return "Invalid"

sales: int = 90
bookshelf_group(sales)
````
